```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data into a pandas dataframe
df2011 = pd.read_csv('2011_rankings.csv')
df2012 = pd.read_csv('2012_rankings.csv')
df2013 = pd.read_csv('2013_rankings.csv')
df2014 = pd.read_csv('2014_rankings.csv')
df2015 = pd.read_csv('2015_rankings.csv')
df2016 = pd.read_csv('2016_rankings.csv')
df2017 = pd.read_csv('2017_rankings.csv')
df2018 = pd.read_csv('2018_rankings.csv')
df2019 = pd.read_csv('2019_rankings.csv')
df2020 = pd.read_csv('2020_rankings.csv')
df2021 = pd.read_csv('2021_rankings.csv')
df2022 = pd.read_csv('2022_rankings.csv')
df2023 = pd.read_csv('2023_rankings.csv')
```

```python
# Combine all the dataframes into a single dataframe
frames = [df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019, df2020, df2021, df2022, df2023]
df = pd.concat(frames)
```

```python
# Count the number of universities per country
country_counts = df['location'].value_counts()
```

#ALL COUNTRIES PIECHART

```python
# Select the top 10 countries with the highest number of universities
top_countries = country_counts.head(10)
```

```python

# Calculate the number of universities in the "Other" category
other_count = country_counts[~country_counts.index.isin(top_countries.index)].sum()
```

```python
# Add the "Other" category to the top countries
top_countries['Other'] = other_count
```

```python

# Print the top 10 countries
print(top_countries)
```

```python
# Print a text table of the top 10 countries
print('{:<7} | {:<20} | {:<10}'.format('sr. no', 'location', 'percentage'))
for i, (country, count) in enumerate(top_countries.items()):
    percentage = count / top_countries.sum() * 100
    print('{:<7} | {:<20} | {:<10.1f}'.format(i+1, country, percentage))

# Plot a pie chart of the top 10 countries
plt.pie(top_countries.values, labels=top_countries.index, autopct='%1.1f%%')
plt.title('Top 10 Countries by Number of Ranked Universities')
plt.show()
```
