```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data into a pandas dataframe

df2020 = pd.read_csv('2020_rankings.csv')
df2021 = pd.read_csv('2021_rankings.csv')
df2022 = pd.read_csv('2022_rankings.csv')
df2023 = pd.read_csv('2023_rankings.csv')
```

```python
# Combine all the dataframes into a single dataframe
frames = [df2020, df2021, df2022, df2023]
df = pd.concat(frames)
```

```python
print(df.columns)
```

```python

plt.scatter(df['stats_student_staff_ratio'], df['scores_research'])
plt.xlabel('Student-Staff Ratio')
plt.ylabel('Research Score')
plt.show()
```

```python
corr = df['stats_student_staff_ratio'].corr(df['scores_research'])
print(corr)
```
