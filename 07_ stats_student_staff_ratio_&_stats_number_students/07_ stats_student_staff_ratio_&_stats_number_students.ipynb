```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data into a pandas dataframe

df2020 = pd.read_csv('2020_rankings.csv')
df2021 = pd.read_csv('2021_rankings.csv')
df2022 = pd.read_csv('2022_rankings.csv')
df2023 = pd.read_csv('2023_rankings.csv')
```

```python
# Combine all the dataframes into a single dataframe
frames = [df2020, df2021, df2022, df2023]
df = pd.concat(frames)
```

```python
# Sort the universities by student-staff ratio
df_sorted = df.sort_values('stats_student_staff_ratio')
```

```python
# Print the top 10 universities with the lowest student-staff ratio
print(df_sorted[['name', 'stats_number_students', 'stats_student_staff_ratio']].head(10))
```

```python
# Print the top 10 universities with the highest number of students
print(df.sort_values('stats_number_students', ascending=False)[['name', 'stats_number_students', 'stats_student_staff_ratio']].head(10))
```

```python
import plotly.express as px

fig = px.scatter(df, x='stats_number_students', y='stats_student_staff_ratio', hover_name='name', log_x=True)
fig.update_layout(title='Number of Students vs. Student-Staff Ratio',
                  xaxis_title='Number of Students (log scale)',
                  yaxis_title='Student-Staff Ratio')
fig.show()
```
