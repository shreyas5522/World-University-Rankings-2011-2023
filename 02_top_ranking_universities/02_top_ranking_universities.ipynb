```python
import pandas as pd
```

```python
# Load the data into a pandas dataframe
df2011 = pd.read_csv('2011_rankings.csv')
df2012 = pd.read_csv('2012_rankings.csv')
df2013 = pd.read_csv('2013_rankings.csv')
df2014 = pd.read_csv('2014_rankings.csv')
df2015 = pd.read_csv('2015_rankings.csv')
df2016 = pd.read_csv('2016_rankings.csv')
df2017 = pd.read_csv('2017_rankings.csv')
df2018 = pd.read_csv('2018_rankings.csv')
df2019 = pd.read_csv('2019_rankings.csv')
df2020 = pd.read_csv('2020_rankings.csv')
df2021 = pd.read_csv('2021_rankings.csv')
df2022 = pd.read_csv('2022_rankings.csv')
df2023 = pd.read_csv('2023_rankings.csv')
```

```python
# Combine all the dataframes into a single dataframe
frames = [df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019, df2020, df2021, df2022, df2023]
df = pd.concat(frames)
```

```python
print(df.columns)
```

```python
# Remove unnecessary columns
df.drop(['aliases', 'closed', 'unaccredited', 'stats_number_students'], axis=1, inplace=True)

# Rename columns for consistency
df.columns = ['rank_order', 'rank', 'name', 'overall_score', 'overall_rank', 'teaching_score', 'teaching_rank', 
              'intl_outlook_score', 'intl_outlook_rank', 'industry_income_score', 'industry_income_rank', 'research_score', 
              'research_rank', 'citations_score', 'citations_rank', 'location', 'subjects_offered', 'student_staff_ratio', 
              'percent_intl_students', 'female_male_ratio']

# Fill in missing values
df.fillna(value={'percent_intl_students': 0, 'female_male_ratio': '0 : 0'}, inplace=True)
```

#TOP 10 UNIVERSITIES | BY OVERALL SCORES

```python
df['overall_score'] = pd.to_numeric(df['overall_score'], errors='coerce')
df[['name', 'overall_score']].sort_values(by='overall_score', ascending=False).head(10)
```
