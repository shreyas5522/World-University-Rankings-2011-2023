```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the data into a pandas dataframe
df2011 = pd.read_csv('2011_rankings.csv')
df2012 = pd.read_csv('2012_rankings.csv')
df2013 = pd.read_csv('2013_rankings.csv')
df2014 = pd.read_csv('2014_rankings.csv')
df2015 = pd.read_csv('2015_rankings.csv')
df2016 = pd.read_csv('2016_rankings.csv')
df2017 = pd.read_csv('2017_rankings.csv')
df2018 = pd.read_csv('2018_rankings.csv')
df2019 = pd.read_csv('2019_rankings.csv')
df2020 = pd.read_csv('2020_rankings.csv')
df2021 = pd.read_csv('2021_rankings.csv')
df2022 = pd.read_csv('2022_rankings.csv')
df2023 = pd.read_csv('2023_rankings.csv')
```

```python
# CombineScores.ipynb all the dataframes into a single dataframe
frames = [df2011, df2012, df2013, df2014, df2015, df2016, df2017, df2018, df2019, df2020, df2021, df2022, df2023]
df = pd.concat(frames)
```

```python
print(df.columns)
```

#Relation Between 'rank' & 'scores_research'

```python
# Convert the 'scores_research' column to a numeric type
df['scores_research'] = pd.to_numeric(df['scores_research'], errors='coerce')
df['rank'] = pd.to_numeric(df['rank'], errors='coerce')
```

```python
# Create a scatter plot to visualize the relationship between scores_research and rank
plt.scatter(df['scores_research'], df['rank'])
plt.xlabel('Scores Research')
plt.ylabel('Rank')
plt.title('Relation between Scores Research and Rank')
plt.show()
```

```python
# Calculate the correlation between scores_research and rank
corr = df['scores_research'].corr(df['rank'])

# Print the correlation in text format
print("The correlation between scores_research and rank is: {:.2f}".format(corr))
```
